```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.pairplot(fifa_df[['Age', 'Overall', 'Potential', 'Stamina', 'Strength']])
```

```python
fifa_df.head()
```

```python
sns.countplot(fifa_df['Age'])
plt.xticks(rotation=90)
plt.title('Age distribution')
plt.show()
```

```python
sns.distplot(fifa_df['Age'])
plt.xticks(rotation=90)
plt.title('Age distribution')
plt.show()
```

```python
countries=fifa_df['Nationality'].value_counts()
index=countries.index
con=pd.DataFrame({'Country':index, 'Count':countries})
con['Percentage%']=(con['Count']/con['Count'].sum())*100
con
```

```python
countries.index
```

```python
con=con['England':'Colombia']
plt.pie(con['Count'],labels=con['Country'],wedgeprops={'linewidth': 3},autopct='%1.1f%%')
plt.title('Distribution of players from top 8 countries')
plt.tight_layout()
plt.show()
```

```python
barca=fifa_df[fifa_df['Club']=='FC Barcelona']
madrid=fifa_df[fifa_df['Club']=='Real Madrid']
print(barca.head())
print(madrid.head())
```

```python
print(barca['Position'].value_counts())
print(madrid['Position'].value_counts())
```

```python
sns.distplot(barca['Age'],color='maroon')
sns.distplot(madrid['Age'],color='black')
plt.title('Comparison of distribution of Age between Barcelona and Real players')
plt.tight_layout()
plt.show()
```
